```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
```

```python
dp = r"C:\Users\anany\Downloads\archive (2)\diabetes.csv"
df = pd.read_csv(dp)
df.head()
```

```python
X=df.drop(['Outcome'], axis=1)
y=df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1) 
```

```python
nbc = GaussianNB()

nbc.fit(X_train, y_train)

y_pred = nbc.predict(X_test)
```

```python
score = nbc.score(X_train, y_train)
score
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
r2 = metrics.r2_score(y_test, y_pred)
r2
```

```python
precision = metrics.precision_score(y_test, y_pred)
precision
```

```python
recall = metrics.recall_score(y_test, y_pred)
recall
```

```python
f1 = metrics.f1_score(y_pred, y_test, average="weighted")
f1
```

```python
labels = ["Not Diabetic", "Diabetic"]
cnf_matrix = metrics.confusion_matrix(y_test, y_pred) 
cnf_matrix 
```

```python
metrics.ConfusionMatrixDisplay(confusion_matrix=cnf_matrix, display_labels=labels).plot()
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(
    n_features=9,
    n_classes=3,
    n_samples=800,
    n_informative=4,
    random_state=1,
    n_clusters_per_class=1,
)
```

```python
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import train_test_split 

from sklearn import metrics 

import pandas as pd 

import numpy as np 

import matplotlib.pyplot as plt 

import seaborn as sns 
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1) 
```

```python
nbc = GaussianNB()

nbc.fit(X_train, y_train)

y_pred = nbc.predict(X_test)
```

```python
score = nbc.score(X_train, y_train)
score
```

```python
r2 = metrics.r2_score(y_test, y_pred)
r2
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
precision = metrics.precision_score(y_test, y_pred, average="weighted")
precision
```
