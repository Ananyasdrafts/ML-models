```python
import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn import preprocessing, svm

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python
dp = r"C:\Users\anany\Downloads\archive\House_Rent_Dataset.csv"
hd = pd.read_csv(dp, index_col="Posted On")
hd.head()
```

```python
X=hd.drop(['Rent'], axis=1)
y=hd['Rent']
```

```python
#find categorical variables
categorical = [var for var in hd.columns if hd[var].dtype=='O']
hd[categorical].head()

```

```python
X = X.drop(['Point of Contact'],axis=1)
```

```python
floors = pd.get_dummies(X['Floor'], drop_first=True)
```

```python
areas = pd.get_dummies(X['Area Type'], drop_first=True)
```

```python
locality = pd.get_dummies(X['Area Locality'], drop_first=True)
```

```python
cities = pd.get_dummies(X['City'], drop_first=True)
```

```python
furnish = pd.get_dummies(X['Furnishing Status'], drop_first=True)
```

```python
tenant = pd.get_dummies(X['Tenant Preferred'], drop_first=True)
```

```python
X = X.drop(['Floor', 'Area Type', 'Area Locality', 'City', 'Furnishing Status', 'Tenant Preferred'],axis=1)
X.head()
```

```python
X = pd.concat([furnish, floors, areas, locality, cities, tenant, X], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)


regr = LinearRegression()

regr.fit(X_train, y_train)

predictions = regr.predict(X_test)

regr.score(X_train, y_train)

```
